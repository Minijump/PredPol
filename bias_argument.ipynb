The following code gives an idea of how much policeman will discriminate the left population in fuction of the parametter bias

```python
import numpy as np
import matplotlib.pyplot as plt
from random import random
```

```python
#bias = 0.4 =>67%
num_iteration = 1000000
```

```python
def get_dis_from_bias(bias):
    cnt = 0
    for i in range(num_iteration):
            if random()*0.001 + bias * random() * 0.001 > random()*0.001:
                cnt+=1
    return(cnt/num_iteration)
```

```python
all_bias = np.arange(0,2,0.05)
all_discrimination = np.array([])
for bias in all_bias:
    all_discrimination = np.append(all_discrimination, get_dis_from_bias(bias))
    
plt.plot(all_bias, all_discrimination)

plt.xlabel('value of bias argument')
plt.ylabel('P policeman will go left for equal P_crime')
plt.title('impact of the bias argument on the policeman distribution')
  
plt.show()
```
